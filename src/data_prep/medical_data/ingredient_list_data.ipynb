```python
df = pd.read_csv('ingredientsList.csv')
```

```python
df.columns
```

```python
cols_to_keep = ['name', 'what_is_it',
       'what_does_it_do', 'who_is_it_good_for', ]
df = df[cols_to_keep]
```

```python
df.dtypes
```

```python
df['who_is_it_good_for'].iloc[0]
```

```python
df['who_is_it_good_for'] = df['who_is_it_good_for'].apply(
    lambda x: [i.strip() for i in ast.literal_eval(x) if i.strip()]
)
```

```python
df
```

```python
df_exploded = df.explode('who_is_it_good_for')
counts = df_exploded['who_is_it_good_for'].value_counts()
counts
```
